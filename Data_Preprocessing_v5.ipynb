## Standardization

One useful data preprocessing technique which is basically used to transform the data attributes with a **Gaussian distribution** (Normal distribution). It differs the mean and SD (Standard Deviation) to a standard Gaussian distribution with a mean of 0 and a SD of 1. This technique is useful in ML algorithms like linear regression, logistic regression that assumes a Gaussian distribution in input dataset and produce better results with rescaled data. We can standardize the data (mean = 0 and SD =1) with the help of `scale` function of scikit-learn preprocessing Python library.

WhatsApp Image 2023-08-27 at 17.46.26.jpeg

```python
# make_blobs dataset example
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
X, y = make_blobs(n_samples=600, n_features=2, centers=4, cluster_std=0.7, random_state=0)
plt.scatter(X[:, 0][y==0], X[:, 1][y==0], s=20)
#plt.scatter(X[:, 0], X[:, 1], s=20)
plt.show()
```

```python
print ("\nIntial data Mean =", X.mean(axis=0))
print ("Initial Std deviation =", X.std(axis=0))
```

```python
# mean removal
X_std = preprocessing.scale(X)
plt.scatter(X_std[:, 0][y==0], X_std[:, 1][y==0], s=20)
plt.show()
```

```python
print ("\nX data standardized Mean =", X_std.mean(axis=0))
print ("X data standardized Std deviation =", X_std.std(axis=0))
```

```python
# Other example of data
data2 = np.array([[ 1, 2,  3, 4],
                 [ 0,  4,  -0.3, 2.1],
                 [ 1,  3.3, -1.9, -4.3],
                 [-1, .1, 3, 2],
                 [.1, .2, .3, .4]])
data2
```

```python
# mean removal
data_standardized = preprocessing.scale(data2)
print(data_standardized)
print ("\ndata_standardized Mean =", data_standardized.mean(axis=0))
print ("data_standardized Std deviation =", data_standardized.std(axis=0))
```

```python
print ("\ndata_standardized Mean =", data_rescaled.mean(axis=0))
print ("data_standardized Std deviation =", data_rescaled.std(axis=0))
```

## Min Max scaling

Most probably our dataset comprises of the attributes with varying scale,
but we cannot provide such data to ML algorithm hence it requires rescaling. Data rescaling makes sure that attributes are at same scale. Generally, attributes are rescaled into the range of 0 and 1. ML algorithms like gradient descent and k-Nearest Neighbors requires scaled data. We can rescale the data with the help of MinMaxScaler class of scikit-learn Python library.


Data Min Max scale at range in `feature_range` parameter

```python
# min max scaling
data_scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
data_scaled = data_scaler.fit_transform(data2)
print ("\nMin max scaled data:\n", data_scaled)
```

```python
print ("\Min max scaled data Mean =", data_scaled.mean(axis=0))
print ("Min max scaled data Std deviation =", data_scaled.std(axis=0))
```

## L1 Normalization

It may be defined as the normalization technique that modifies the dataset values in a way that **in each row** the sum of the absolute values will always be up to 1. It is also called **Least Absolute Deviations**.

```python
data2
```

## L2 Normalization
It may be defined as the normalization technique that modifies the
dataset values in a way that in each row **the sum of the squares** will always be up to 1. It is also called least squares.

```python
# L2 normalization
data_normalized_l2 = preprocessing.normalize(data2, norm='l2')
print ("L2 normalized data:\n", data_normalized_l2)
```

```python
# square computation
sq=np.square(data_normalized_l2)
sq
```

```python
# least squares
for i in range(5):
  print(sum(sq[i]))
```

```python
# other way to sum the rows
sq.sum(axis=1)
```

## Binarization

As the name suggests, this is the technique with the help of which we can make our data binary. We can use a binary threshold for making our data binary. The values above that threshold value will be converted to 1 and below that threshold will be converted to 0. For example, if we choose threshold value = 0.5, then the dataset value above it will become 1 and below this will become 0. That is why we can call it binarizing the data or thresholding the data. This technique is useful when we have probabilities in our dataset and want to convert them into crisp values.

We can binarize the data with the help of Binarizer class of scikit-learn Python library.

```python
# binarization
data_binarized = preprocessing.Binarizer(threshold=1.4).transform(data2)
print ("\nBinarized data:\n", data_binarized)
```

## One Hot Encoding
A one hot encoding is a representation of categorical variables as binary vectors.

This first requires that the categorical values be mapped to integer values.

Then, each integer value is represented as a binary vector that is all zero values except the index of the integer, which is marked with a 1.

```python
# one hot encoding
encoder = preprocessing.OneHotEncoder()
x=np.array([[0], [2], [1], [3], [7]])
print(x)
encoder.fit(x)
encoded_vector = encoder.transform(x).toarray()
print ("\nEncoded vector:\n", encoded_vector)
```

In this example, we will use the encoders from the scikit-learn library. Specifically, the `LabelEncoder` of creating an integer encoding of labels and the `OneHotEncoder` for creating a one hot encoding of integer encoded values.

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
# define example
data3 = ['cold', 'cold', 'warm', 'cold', 'hot', 'hot', 'warm', 'cold', 'warm', 'hot']
values = np.array(data3)
print(values)
```

```python
# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(values)
print(integer_encoded)
```

```python
integer_encoded_reshaped = integer_encoded.reshape(len(integer_encoded), 1)
integer_encoded_reshaped
```

```python
# binary encode
onehot_encoder = OneHotEncoder(sparse_output=False)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded_reshaped)
print(onehot_encoded)
```

```python
# invert first line example
inverted = label_encoder.inverse_transform([np.argmax(onehot_encoded[0, :])])
print(inverted)
```

```python
# Exemplo de pr-processamento
# Contribuio de Gabriel Delongui Polvani <gabriel.polvani@uel.br>

import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn import preprocessing

X, Y = make_blobs(n_samples=400, centers=2, cluster_std=3.00, random_state=1)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(ncols=2, nrows=2, tight_layout=True, sharex=True, sharey=True)

# Normalizao normal-Padro
X_normal = preprocessing.scale(X)
ax1.scatter(*X_normal[Y==0].T, color='#12629A', alpha=0.6, marker='o')
ax1.scatter(*X_normal[Y==1].T, color='#F247B2', alpha=0.6, marker='x')
ax1.legend(['Classe 1', 'Classe 2'], loc='lower right')

# Normalizao MinMax
X_MinMax = preprocessing.MinMaxScaler(feature_range=(0, 1))
X_MinMax = X_MinMax.fit_transform(X)

ax2.scatter(*X_MinMax[Y==0].T, color='#12629A', alpha=0.6, marker='o')
ax2.scatter(*X_MinMax[Y==1].T, color='#F247B2', alpha=0.6, marker='x')
ax2.legend(['Classe 1', 'Classe 2'], loc='lower right')

# Normalizao L1
X_L1 = preprocessing.normalize(X, norm='l1')

ax3.scatter(*X_L1[Y==0].T, color='#12629A', alpha=0.6, marker='o')
ax3.scatter(*X_L1[Y==1].T, color='#F247B2', alpha=0.6, marker='x')
ax3.legend(['Classe 1', 'Classe 2'], loc='lower right')

# Normalizao L2
X_L2 = preprocessing.normalize(X, norm='l2')

ax4.scatter(*X_L2[Y==0].T, color='#12629A', alpha=0.6, marker='o')
ax4.scatter(*X_L2[Y==1].T, color='#F247B2', alpha=0.6, marker='x')
ax4.legend(['Classe 1', 'Classe 2'], loc='lower right')

ax1.grid(alpha=0.6), ax2.grid(alpha=0.6), ax3.grid(alpha=0.6), ax4.grid(alpha=0.6)

# Adicionando texto para as legendas dos subplots e legendas dos eixos (Coordenadas x e y obtidas empiricamente)
fig.text(0.075, 0.94, 'Standard Normal')
fig.text(0.548, 0.94, 'MinMax (0,1)')
fig.text(0.075, 0.47, 'L1 Normalization')
fig.text(0.548, 0.47, 'Standard Normal')

fig.text(0.51, 0, '$x_1$')
fig.text(0, 0.52, '$x_2$', rotation=90)

plt.show()
```

```python
# Contribuio de Joo Paulo Monteiro Santana <joao.paulo.santana@uel.br>

# Algoritmo KNN para os seguintes datasets: X, X_normal, X_MinMax, X_L1 e X_L2.
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn import preprocessing

X, Y = make_blobs(n_samples=400, centers=2, cluster_std=3.00, random_state=1)

# Subdiviso entre as etapas treinamento e teste
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train);
Y_pred = knn.predict(X_test)

# Plot para o banco de dados padro

fig, axs = plt.subplots(ncols=3, nrows=1, figsize=(9,3), tight_layout=True, sharey=True)

axs[0].scatter(X_train[:,0],X_train[:,1],c=Y_train)
axs[0].set_title('Training Set')
axs[0].grid()

axs[1].scatter(X_test[:,0],X_test[:,1],c=Y_test)
axs[1].set_title('Test Set')
axs[1].grid()

axs[2].scatter(X_test[:,0],X_test[:,1],c=Y_pred)
axs[2].set_title('Predict Set')
axs[2].grid()

plt.show()
```
